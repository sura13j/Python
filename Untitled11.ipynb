```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup = BeautifulSoup(soup.text, 'lxml')
soup
```

```python
table = soup.table
table.find_all('tr')
table
```

```python
data = []
for tables in table:
    x = data.append(table.text.split('\n'))
    if x[1] != "":
        data.append(x)
data
```

```python
import csv
file = open("covid.csv", 'w')
y = csv.writer(file)
y.writerows(data)
file.close()
```
